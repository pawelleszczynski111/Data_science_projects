```python
 import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('wine.csv')
```

```python
df.head(5)
```

```python
print(df.shape)
```

```python
X= df.drop(columns=['Wine'],axis=1)
Y=df['Wine']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=7)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
LinearRegression()
print('\nCoefficient of model :', model.coef_)
```

```python
print('\nIntercept model', model.intercept_)
```

```python
predict_train=model.predict(x_train)
print('\nItem on train data',predict_train)
```

```python
rmse_train=np.sqrt(mean_squared_error(y_train,predict_train))
print('\nRMSE on train dataset : ', rmse_train)
```

```python
predict_test=model.predict(x_test)
print('\nItem on test data',predict_test)
```

```python
rmse_test=np.sqrt(mean_squared_error(y_test,predict_test))
print('\nRMSE on test dataset : ', rmse_test)
```
